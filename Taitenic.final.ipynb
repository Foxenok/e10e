```python
#Какое самое популярное женское имя на корабле? Извлеките из
#полного имени пассажира (колонка Name) его личное имя (First
#Name). Это задание — типичный пример того, с чем сталкивается специалист по анализу данных
#Vestrom, Miss. Hulda Amanda Adolfina
def col(S):
    if S.find('Mr.') >= 0:
        return
    elif S.find('Miss. ') >= 0:
        ss = S.find('Miss. ')
        S = S[ss+6:]
        cc = S.find(' ')
        
        return S[:cc]
        
    else:
        rs = S.find('(')


        S = S[rs+1:]
        sr = S.find(' ')
        d = S[:sr]
        if d[-1] == ',':
            return d[:-1]
        else:
        
            return d
```

```python
def sol(s):
    if s.find('female')>=0:
        return 1
    else:
        return 0
```
